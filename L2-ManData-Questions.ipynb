**Question 1:**

*The data below is the income statements of GlaxoSmithKline from Q1 2016 to Q2 2020. [1]*

*Produce a pivot table from the `gsk` DataFrame with an index column of `Year` which calculates the sum of the `Turnover` and `Cost of sales`. You will need to use the argument `aggfunc = sum` to calculate the sum.*

*Save this pivot table as a variable labelled `gsk_yearly`.*

```python
import pandas as pd 
gsk = pd.read_csv("https://raw.githubusercontent.com/ThomasJewson/datasets/master/GSK-QuarterlyIncomeStatements2016-2020.csv")
gsk
```

```python
# Solution
gsk_yearly = gsk.pivot_table(index="Year",aggfunc=sum)
gsk_yearly
```

**Question 2:**

*Calculate the yearly gross profit of GSK in the `gsk_yearly` DataFrame, outputting the answers into a new column labelled `Gross profits`.*

```python
#Solution
gsk_yearly["Gross profits"] = gsk_yearly["Turnover"] - gsk_yearly["Cost of sales"]
gsk_yearly
```

**Question 3:**

Due to the COVID-19 outbreak the first and second quarters of 2020 might be anomalous. Replace `2020` in the `gsk` DataFrame with `"2020*"` using the `.replace()` function or the `.loc[]` method. 

```python
#Solution
gsk.loc[gsk["Year"] == 2020,"Year"] = "2020*"
```

```python
gsk.replace(2020,"2020*")
```

**Question 4:**

*From the `PrisonDF` below, calculate the percentage full of each prison and save the data to a new column within the DataFrame called `Percentage Full`. Then, sort `Percentage Full` from largest to smallest.*

```python
PrisonDF = pd.read_csv("https://raw.githubusercontent.com/ThomasJewson/datasets/master/PrisonsDataFull.csv")
PrisonDF
```

```python
#Solution
PrisonDF["Percentage Full"] = (PrisonDF["Population"] / PrisonDF["Capacity"])*100
PrisonDF.sort_values("Percentage Full",ascending=False)
```

**Question 5:**

*The `CheeseDF` DataFrame has sales data from an imaginary cheese shop. Produce a pivot table from `CheeseDF` which firsts pivots about the type of cheese then pivots about the quarter, producing a sum of just the `Units Sold` column.*

```python
CheeseDF = pd.read_csv("https://raw.githubusercontent.com/ThomasJewson/datasets/master/CheeseShop.csv")
CheeseDF
```

```python
CheeseDF.pivot_table(index=["Cheese","Quarter"],aggfunc=sum,values="Units Sold")
```
